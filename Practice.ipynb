Draw a circle of radius 5 using Matplotlib using a scatter plot.


```python
r=5
x_val= np.linspace(-r,r,1000000)
X_val= np.linspace(-r,r,1000000)
y_val=[]
Y_val=[]
for x in x_val:                          #upper half
  y_val.append(((r**2)-(x**2))**0.5)
for x in X_val:                         #lower half
  Y_val.append(-(((r**2)-(x**2))**0.5))
plt.scatter(x_val,y_val)
plt.scatter(X_val,Y_val)
```

Perform the following experiment:

(1) Throw a standard, unbiased, six-sided dice nn times.

(2) Get the frequency of occurrence of the numbers from 11 to 66.

(3) Convert the frequencies to probabilities.

(4) Represent the probabilities of obtaining each of the six numbers in the form of a bar plot. You can use plt.bar for this purpose.

Run this experiment for the following values of nn: 10, 100, 1000, 10000, 100000, 100000010,100,1000,10000,100000,1000000 and note down your observations.

```python
n=1000000
occur=[]
l=[]
bins=[1,2,3,4,5,6] 
for i in range(n):
  occur.append(np.random.randint(1,7))
d={}
for x in bins:
  d[x]=0
for y in occur:
  d[y]+=1
for z in d:
  l.append(d[z]/n) 
plt.bar(bins,l)
    
    
```
